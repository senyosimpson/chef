```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def gaussian_basis(x, u, s):
    x = np.exp(-(x-u)**2/(2*s**2))
    return x
    
def sigmoidal_basis(x, u, s):
    x = 1/(1 + np.exp(-(x-u)/s))
    return x

def identity_basis(x):
    return x
```

```python
def linear_fn(x, a=0.5, b=0.3):
    fn = a * x + b
    fn_noise = fn + np.random.randn(len(x))
    return fn, fn_noise
```

```python
# Generate fake data
steps = 10
x = np.arange(steps)
y, t = linear_fn(x)
plt.plot(x, y)
plt.scatter(x, y, s=10, c='r')
plt.plot(x, t)
plt.scatter(x, t, s=10, c='r')
```

```python
M = 2
x = x.reshape(-1, 1)
# Design Matrix is denoted by X
bias = np.ones(steps).reshape(-1, 1)
X = np.repeat(x, M-1, axis=1)
X = np.concatenate((bias, X), axis=1)
X
```

```python
w = np.matmul(X.T, X)
w = np.linalg.inv(w)
w = np.matmul(w, X.T)
w = np.matmul(w, t)
w = w.reshape(-1, 1)
w
```

```python
t_hat = np.matmul(w.T, X.T).T
t_hat
```

```python
plt.figure(figsize=(20, 5))
plt.plot(x, y, label='Original Line')
plt.scatter(x, t, s=40, c='blue')
plt.plot(x, t_hat, label='Best Fit Line')
plt.scatter(x, t_hat, s=40, c='k')
plt.legend()
```

```python
x = 2 * np.random.rand(100, 1)
y = 4 + 3 * x
t = y + np.random.randn(100, 1)

plt.scatter(x, y, s=8)
plt.scatter(x, t, s=8)
```

```python
# init weight parameters
w = np.random.randn(2, 1)
print('Init Weights:', w.flatten())
X = np.hstack((np.ones(x.shape), x)).T
learning_rate = 0.01

for _ in range(10):
    for i in range(10):
        start = i*10
        end = start + 10
        x_ = X[:, start:end]
        t_ = t[start:end]
        loss = t_ - np.matmul(w.T, x_).T
        loss = np.matmul(loss.T, x_.T).T
        w = w + learning_rate * loss
        
print('Weights after training:', w.flatten())
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(x, t, s=8)
t_hat = np.matmul(w.T, X).reshape(-1, 1)
plt.plot(x, t_hat)
```
